```python
from pathlib import Path
import json
import pandas as pd

PROJECT_ROOT = Path.cwd().parent
```

```python
annotation_path = (
    PROJECT_ROOT
    / "data"
    / "raw"
    / "fashionpedia"
    / "annotations"
    / "instances_attributes_val2020.json"
)

with open(annotation_path, "r") as f:
    data = json.load(f)

print(type(data))
print(data.keys())
```

```python
print("Images      :", len(data["images"]))
print("Annotations :", len(data["annotations"]))
print("Categories  :", len(data["categories"]))
print("Attributes  :", len(data["attributes"]))
```

```python
category_lookup = {
    c["id"]: c["name"]
    for c in data["categories"]
}

attribute_lookup = {
    a["id"]: a["name"]
    for a in data["attributes"]
}
```

```python
ann = data["annotations"][0]

print("Category:")
print(category_lookup[ann["category_id"]])

print()

print("Attributes:")

for attr in ann["attribute_ids"]:
    print(attribute_lookup[attr])
```
