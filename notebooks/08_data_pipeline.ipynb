```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT))
```

```python
from src.data.scanner import DatasetScanner
from src.data.label_encoder import LabelEncoder
from src.data.dataset import ClothingDataset
from src.data.transforms import get_train_transforms
from src.data.dataloader import create_dataloader
```

```python
scanner = DatasetScanner("../data/sample")

image_paths, labels, classes = scanner.scan()

print(classes)
```

```python
encoder = LabelEncoder(classes)

encoded_labels = [
    encoder.encode(label)
    for label in labels
]

print(encoded_labels[:10])
```

```python
dataset = ClothingDataset(
    image_paths=image_paths,
    labels=encoded_labels,
    transform=get_train_transforms()
)

print(len(dataset))
```

```python
image, label = dataset[0]

print(image.shape)
print(label)
```

```python
dataloader = create_dataloader(
    dataset,
    batch_size=4,
    shuffle=True
)

images, labels = next(iter(dataloader))

print(images.shape)
print(labels)
```
