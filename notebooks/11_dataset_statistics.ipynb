```python
from collections import Counter

category_counter = Counter()

for ann in adapter.data["annotations"]:
    category = adapter.category_lookup[ann["category_id"]]
    category_counter[category] += 1

print(category_counter.most_common(20))
```

```python
attribute_counter = Counter()

for ann in adapter.data["annotations"]:
    for attr in ann["attribute_ids"]:
        attribute_counter[adapter.attribute_lookup[attr]] += 1

print(attribute_counter.most_common(20))
```

```python
from collections import defaultdict

image_counter = defaultdict(int)

for ann in adapter.data["annotations"]:
    image_counter[ann["image_id"]] += 1

print("Average garments per image:",
      sum(image_counter.values()) / len(image_counter))
```
